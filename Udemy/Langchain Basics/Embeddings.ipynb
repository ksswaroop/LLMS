```python
## Reading PDF
from langchain_community.document_loaders import PyPDFLoader
loader=PyPDFLoader("/home/swaroop/Documents/Coding/LLMS/Udemy/Material/1-Basics+Of+Langchain/3.2-DataIngestion/attention.pdf")
pdf_doc=loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_doc=text_splitter.split_documents(pdf_doc)
```

```python
final_doc
```

```python
## Vector Embedding an d VectorStoreDB -- Chroma
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_doc,embeddings)
db
```

```python
query="How does the Transformer handle the order of words in a sequence?"
retrieve=db.similarity_search(query)
print(retrieve)
```
